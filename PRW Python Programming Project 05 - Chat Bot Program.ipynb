```python
print("Jack: Hello I'm jack.What is your name?")
name=input("Your answer :")
print(f"Jack: Hi {name}... So what do you want to know about our company?")
ans=input("Your answer :")

def end_chat(user_name):
    global flag,ans
    check=input("Your answer :")
    if any(i in check.lower().split(" ") for i in ["yes","yeah"]):
        print("Jack: Okay. What do you want to know?")
        ans=input("Your answer :")
        flag=True
    else:
        print(f"Jack: Nice to chat with you {user_name}. Have a nice day...!!")
        flag=False

flag=True

while flag:
    if all(i in ans.lower().split(" ") for i in ["company","name"]):
        print("Jack: Our company is Leeza(PVT)Ltd")
        print("Jack: Do you want to know any thing else?")
        end_chat(name)
            
    elif all(i in ans.lower().split(" ") for i in ["ceo","name"]):
        print("Jack: Our CEO is Mr. James")
        print("Jack: Do you want to know any thing else?")
        end_chat(name)
            
    elif all(i in ans.lower().split(" ") for i in ["address"]):
        print("Jack: 125 Srimath Anagarika Dharmapala Mawatha, Colombo 7.")
        print("Jack: Do you want to know any thing else?")
        end_chat(name)
            
    elif any(i in ans.lower().split(" ") for i in ["contact","email","telephone","mobile"]):
        print("Jack: Our email is xyz@gmail.com and contact number is 011 2111111")
        print("Jack: Do you want to know any thing else?")
        end_chat(name)
    
    else:
        print(f"Jack: Your question is not clear {name}. Do you want to know any thing else?")
        end_chat(name)   
```

# Voice Chatbot

```python
import pyttsx3
eng=pyttsx3.init()
eng.setProperty("rate",110)


eng.say("Hello I'm jack. What is your name?")
eng.runAndWait()
name=input("Your answer :")

eng.say(f"Hi {name}... So what do you want to know about our company?")
eng.runAndWait()
ans=input("Your answer :")

def end_chat(user_name):
    global flag,ans
    check=input("Your answer :")
    if any(i in check.lower().split(" ") for i in["yes","yeah"]):
        eng.say("What do you want to know?")
        eng.runAndWait()
        ans=input("Your answer :")
        flag=True
    else:
        eng.say(f"Nice to chat with you {user_name}. Have a nice day...!!")
        eng.runAndWait()
        flag=False

flag=True

while flag:
    if all(i in ans.lower().split(" ") for i in["company","name"]):
        eng.say("Our company is Leeza Private Limited")
        eng.runAndWait()
        
        eng.say("Do you want to know any thing else?")
        eng.runAndWait()

        end_chat(name)
            
    elif all(i in ans.lower().split(" ") for i in["ceo","name"]):
        eng.say("Our CEO is Mr. James")
        eng.runAndWait()
        
        eng.say("Do you want to know any thing else?")
        eng.runAndWait()
        
        end_chat(name)
            
    elif all(i in ans.lower().split(" ") for i in["address"]):
        eng.say("125 Srimath Anagarika Dharmapala Mawatha, Colombo 7.")
        eng.runAndWait()

        eng.say("Do you want to know any thing else?")
        eng.runAndWait()
        
        end_chat(name)
            
    elif any(i in ans.lower().split(" ") for i in["contact","email","telephone","mobile"]):
        eng.say("Our email is xyz@gmail.com and contact number is 011 2111111")
        eng.runAndWait()

        eng.say("Do you want to know any thing else?")
        eng.runAndWait()
        
        end_chat(name)
    
    else:
        eng.say(f"Your question is not clear {name}. Do you want to know any thing else?")
        eng.runAndWait()
        
        end_chat(name)   
```
